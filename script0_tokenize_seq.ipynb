```python
# helper function to make sure all the sequence name have a consistant form

def check_modification_cnt(seq):
    
    plus_sign_cnt = 0
    for i in range(len(seq)):
        if seq[i] == "+":
            plus_sign_cnt = plus_sign_cnt + 1
            
    return plus_sign_cnt
            
```

```python
cnt = check_modification_cnt('abc+xx+abc')
print(cnt)
```

```python
# output is expected to be zero
multiple_modi = 0
for s in df['Name']:
    if check_modification_cnt(s) > 1:
        multiple_modi = multiple_modi + 1
        
print(multiple_modi)
    
    
```

```python
# two rows that don't have modifications
non_modi = 0
for s in df['Name']:
    if check_modification_cnt(s) == 0:
        non_modi = non_modi + 1
        
print(non_modi)
```

```python
# helper function to tokenize seq 
# input : iedb_seq, e.q. GLLGFVFTL + MCM(G1, L2)
# output: list that contains all the tokens from input, ['MCM_G', 'MCM_L', 'L', 'G', 'F', 'V', 'F', 'T', 'L']
# tokens that has "_" represents its a ncaa
def tokenize_seq(iedb_seq):
    iedb_seq = iedb_seq.replace(" ", "")
    token_list = []
    
    pos_of_plus = iedb_seq.rfind('+')
    pos_of_par1 = iedb_seq.rfind('(')
    pos_of_par2 = iedb_seq.rfind(')')
    
    if pos_of_plus == -1:
        for i in range(len(iedb_seq)):
            token_list.append(iedb_seq[i])
        return token_list
    
    else:
        caa_seq = iedb_seq[0:pos_of_plus]
        aa_dict = {}
    
        for i in range(pos_of_plus):
            aa_dict[i + 1] = iedb_seq[i]
        
        
        ncaa_seq = iedb_seq[(pos_of_par1 + 1) : pos_of_par2]
        modif_method = iedb_seq[(pos_of_plus + 1) : pos_of_par1]
    
        ncaas = ncaa_seq.split(",")
        ncaa_dict = {}
    
        for i in range(len(ncaas)):
            ncaa_token = ncaas[i]
            ncaa_index = ncaa_token[1:]
            ncaa = ncaa_token[0]
            ncaa_dict[int(ncaa_index)] = modif_method + "_" + ncaa
        
        
        for key in ncaa_dict:
            aa_dict[key] = ncaa_dict[key]
        
        return list(aa_dict.values())
        
    
```

```python
# test1 of helper function tokenize_seq(iedb_seq)
iedb_seq1 = "GLLGFVFTL + MCM(G1, L2)"

seq_list1 = tokenize_seq(iedb_seq1)
```

```python
# apply tokenize_seq to column of 'name' so to generate a new column named 'token_list'
df['token_list'] = df['Name'].apply(tokenize_seq)
```

```python
def has_token(token_list, token = "PYRE_Q"):
    for t in token_list:
        if token in t:
            return True
        
    return False
```

```python
df['has_token'] = df['token_list'].apply(has_token)
```
