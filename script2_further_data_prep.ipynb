```python
def tokenize_seq(iedb_seq):
    iedb_seq = iedb_seq.upper()
    iedb_seq = iedb_seq.replace(" ", "")
    token_list = []
    
    pos_of_plus = iedb_seq.rfind('+')
    pos_of_bracket = iedb_seq.rfind('[')
    pos_of_par1 = iedb_seq.rfind('(')
    pos_of_par2 = iedb_seq.rfind(')')
    
    if pos_of_plus == -1:
        if pos_of_bracket == -1:
            
            for i in range(len(iedb_seq)):
                token_list.append(iedb_seq[i])
            return token_list
        else :
            return tokenize_seq_1(iedb_seq)
    
    else:
        caa_seq = iedb_seq[0:pos_of_plus]
        aa_dict = {}
    
        for i in range(pos_of_plus):
            aa_dict[i + 1] = iedb_seq[i]
        
        
        ncaa_seq = iedb_seq[(pos_of_par1 + 1) : pos_of_par2]
        modif_method = iedb_seq[(pos_of_plus + 1) : pos_of_par1]
    
        ncaas = ncaa_seq.split(",")
        ncaa_dict = {}
    
        for i in range(len(ncaas)):
            ncaa_token = ncaas[i]
            ncaa_index = ncaa_token[1:]
            ncaa = ncaa_token[0]
            ncaa_dict[int(ncaa_index)] = modif_method + "_" + ncaa
        
        
        for key in ncaa_dict:
            aa_dict[key] = ncaa_dict[key]
        
        return list(aa_dict.values())
        
    
```

```python
def tokenize_seq_1(seq):
    token_list = []
    into_bracket = 0
    i = 0
    while i < len(seq):
        c = seq[i]
        start = i
        if c == '[':
            
            while i < len(seq) and seq[i] != ']' :
                #print(seq)
                i = i + 1              
            t = seq[start +1 : i]  
            token_list.append(t)
            
            
        else:
            token_list.append(seq[i])
            
        i = i +1 
        
    return token_list
```
